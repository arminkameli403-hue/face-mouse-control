```python
import cv2
import pyautogui as robot
```

```python
face_model = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_model = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
cam = cv2.VideoCapture(0) 
```

```python
while True:
    _, img = cam.read()
    img = cv2.flip(img,1)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face = face_model.detectMultiScale(gray, 1.1, 5)

    if (len(face)>0):
        x = face[0][0]
        y = face[0][1]
        x2 = x + face[0][2]
        y2 = y + face[0][3]


        gray_face = gray[y:y2, x:x2]
        eyes = eye_model.detectMultiScale(gray_face, 1.1, 5, minSize=(30, 30))
        
        white = (250,250,250)
        red = (0,0,250)
        rang = white
        
        mouse_x, mouse_y = robot.position()
        
        if y < 80:
            rang=red
            mouse_y = mouse_y - abs (y-80)
            robot.moveTo(mouse_x , mouse_y,0)

        if y2 > 300:
            rang=red
            mouse_y = mouse_y + abs (y2-350)
            robot.moveTo(mouse_x , mouse_y,0.2)


        
        if x< 200:
            rang = red
            mouse_x=mouse_x - 10
            robot.moveTo(mouse_x,mouse_y,0.2)
        
        if x2 > 450:
            rang = red
            mouse_x=mouse_x + 10
            robot.moveTo(mouse_x,mouse_y,0.2)
        out =cv2.rectangle(img, (x, y), (x2, y2), (0, 255, 0), 3)
        out = cv2.rectangle(img,(200,80),(450,300),rang,3)

        
            
        print(f'x: {x}, y: {y}')

    cv2.imshow('Face and Eye Detection', img)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
